```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing


import  re
pd.set_option('display.max_columns', 50)
```

Metabolomics

```python
#Read Data
Metabolomics_original = pd.read_csv("../input/matrix-metabolomics-pca/matrix_metabolomics_pca.csv",header=None)
metabolomics = Metabolomics_original.copy()
print(metabolomics.shape)
metabolomics.head()
```

```python
# Remove model_id column (first column)
first_column = metabolomics.columns[0]
metabolomics = metabolomics.drop([first_column], axis=1)
metabolomics.head()
```

```python
X = metabolomics.iloc [:,0:225].values
print(X.shape)
```

```python
# Center and scale the data
scaled_data = preprocessing.scale(X)
```

```python
#Setting 90% variance 
pca = PCA (n_components =0.90, whiten = True)
Meta1_pca = pca.fit_transform(scaled_data)

print('Original no. of features:', X.shape[1])
print('Reduced no. of features:', Meta1_pca.shape[1])
```

```python
#Setting 95% variance (Experimental purposes)
pca = PCA (n_components =0.95, whiten = True)
Meta2_pca = pca.fit_transform(scaled_data)

print('Original no. of features:', X.shape[1])
print('Reduced no. of features:', Meta2_pca.shape[1])
```

```python
print(Meta1_pca.shape)
```

```python
pd.DataFrame(Meta1_pca).to_csv("./Meta_pca.csv", header=None, index=None)
```

```python
#Visualise no. of componenets vs cum. explained variance
pca = PCA().fit(scaled_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
#To visualise Scree Plot

#Create PCA object
pca = PCA()
#Calculating loading scores and the varation each principle component accounts for
pca.fit(scaled_data)
#Generate coordinates for a PCA graph based on loading scores and the scaled data
pca_data = pca.transform(scaled_data)
```

```python
# Draw a scree plot and a PCA plot

#/Calculate the percentage of variation that each principal component accounts for 
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
#/Create labels
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
print(per_var)
#/Create Bar Plot
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')

fig = plt.rcParams["figure.figsize"] = (40,100)
plt.show()
plt.savefig("./meta_pca.png")
```
