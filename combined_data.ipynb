```python
rppa_index = pd.read_csv('../input/dnnrppa/DNN_Input_RPPA.csv', usecols= ['model_id'])
miRNA_index = pd.read_csv('../input/dnnmirna/DNN_Input_miRNA.csv', usecols = ['model_id'])
meta_index = pd.read_csv('../input/dnn-meta/DNN_Input_Metabolomics.csv', usecols = ['model_id'])
exp_index = pd.read_csv('../input/dnnexp/DNN_Input_Expression.csv', usecols = ['model_id'])
mut_index = pd.read_csv('../input/dnnmutations/DNN_Input_Mutation.csv', usecols = ['model_id'])
cnv_index = pd.read_csv('../input/dnncnv/DNN_Input_CNV.csv', usecols = ['model_id'])
```

```python
#Filtering common indexes in 6 files
common1 = pd.merge(rppa_index,miRNA_index, on="model_id")
common2 = pd.merge(common1,meta_index, on="model_id")
common3 = pd.merge(common2,exp_index, on="model_id")
common4 = pd.merge(common3,mut_index, on="model_id")
common_index_six_data = pd.merge(common4,cnv_index, on="model_id")
```
