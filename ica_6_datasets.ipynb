Metabolomics

```python
#Read Data
Metabolomics_original = pd.read_csv("../input/matrix-metabolomics-pca/matrix_metabolomics_pca.csv",header=None)
metabolomics = Metabolomics_original.copy()
print(metabolomics.shape)
metabolomics.head()
```

```python
# Remove model_id column (first column)
first_column = metabolomics.columns[0]
metabolomics = metabolomics.drop([first_column], axis=1)
metabolomics.head()
```

```python
X = metabolomics.iloc [:,0:225].values
print(X.shape)
```

```python
from sklearn.decomposition import FastICA

transformer_Meta = FastICA(n_components=80,random_state=0,tol=0.0001,max_iter=500)
X_transformed_Meta = transformer_Meta.fit_transform(X)
X_transformed_Meta.shape
```

```python
pd.DataFrame(X_transformed_Meta).to_csv("./Meta_Fast_ICA.csv", header=None, index=None)
```

```python
CNV_original = pd.read_csv("../input/matrix-cnv-pca-1/matrix_cnv_pca (1).csv",header=None)
cnv = CNV_original.copy()
print(cnv.shape)
cnv.head()
```

```python
X = cnv.iloc [:,0:24500].values
print(X.shape)
```

```python
scaled_data = preprocessing.scale(X)
```

```python
#Imputing Values (https://leandeep.com/datalab-kaggle/handling-missing-values.html)
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
data_with_imputed_values = my_imputer.fit_transform(scaled_data)
print(data_with_imputed_values)
```

```python
transformer_CNV = FastICA(n_components=88,random_state=0,tol=0.0001,max_iter=500)
X_transformed_CNV = transformer_CNV.fit_transform(data_with_imputed_values)
X_transformed_CNV.shape
```

```python
pd.DataFrame(X_transformed_CNV).to_csv("./CNV_Fast_ICA.csv", header=None, index=None)
```
