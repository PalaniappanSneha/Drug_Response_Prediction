```python
#Drug Response Data Prep
drug_original2 = pd.read_csv(r'../input/drug-response/matrix_drug_response_S.csv')
print(drug_original2.shape)
drug_original2.set_index('model_id',inplace=True)
drug2 = drug_original2.copy()
drug_original2.head()
```

```python
#Remove columns where 95% of the data is 0. This is done to select the top drugs for testing. 
drug2 = drug2.loc[:,(drug2==0).sum()<len(drug2)*0.050]
print(drug2.shape)
print('no of samples tested on all drugs:',((drug2.transpose()==0).sum()==0).sum())
```

```python
drug2.columns
```

```python
#Matrix_Metabolomics (before PCA)
metabolomics_original1 = pd.read_csv('../input/original-meta/matrix_metabolomics_S.csv')
id_index = metabolomics_original1['model_id']
metabolomics_original1.shape
metabolomics_original1.head()
```

```python
#Appending model_id from Matrix_Metabolomics to PCA output data

metabolomics_original2 = pd.read_csv(r'../input/meta-pca/Metabolomics_PCA.csv',header=None)
#print(metabolomics_original2.shape)
#metabolomics_original2.head()

metabolomics_original2 = metabolomics_original2.join(id_index)

cols =list(metabolomics_original2.columns)
cols = [cols[-1]] + cols[:-1]
metabolomics_original2 = metabolomics_original2[cols]

metabolomics_original2.head()
```

```python
print(metabolomics_original2.shape)
metabolomics_original2.set_index('model_id',inplace=True)
metabolomics_original2.head()
```

```python
metabolomics2 = metabolomics_original2.copy()
print(metabolomics2.shape)
```

```python
# last columns is Y
metabolomics2 = metabolomics2.merge(drug2,
                 how = 'left',
                right_index = True,
                left_index = True)

metabolomics2.dropna(axis='rows', inplace = True)
print(metabolomics2.shape)
```

```python
#No PCA Metabolomics Data
print(metabolomics_original1.shape)
metabolomics_original1.set_index('model_id',inplace=True)
metabolomics_original1.head()
```

```python
metabolomics1 = metabolomics_original1.copy()
print(metabolomics1.shape)
```

```python
# last columns is Y
metabolomics1 = metabolomics1.merge(drug2,
                 how = 'left',
                right_index = True,
                left_index = True)

metabolomics1.dropna(axis='rows', inplace = True)
print(metabolomics1.shape)
```
